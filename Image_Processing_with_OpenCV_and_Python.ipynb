```python
import pandas as pd
import numpy as np
from glob import glob
import cv2
import matplotlib.pyplot as plt
import os

# Adjusted target directories based on actual upload path
cat_dir = '/content/test_set/test_set/cats/'
dog_dir = '/content/test_set/test_set/dogs/'

# Verify if the target directories exist
cat_dir_exists = os.path.isdir(cat_dir)
dog_dir_exists = os.path.isdir(dog_dir)

print("Cat directory exists:", cat_dir_exists)
print("Dog directory exists:", dog_dir_exists)

# List files in the cat directory if it exists
if cat_dir_exists:
    cat_files_listed = os.listdir(cat_dir)
    print("Files in cat directory:", cat_files_listed[:5])  # Print first 5 files for verification
else:
    print("Cat directory not found")

# List files in the dog directory if it exists
if dog_dir_exists:
    dog_files_listed = os.listdir(dog_dir)
    print("Files in dog directory:", dog_files_listed[:5])  # Print first 5 files for verification
else:
    print("Dog directory not found")

# Retrieve the list of cat and dog image files using glob
cat_files = glob(os.path.join(cat_dir, '*.jpg'))
dog_files = glob(os.path.join(dog_dir, '*.jpg'))

# Check the length and contents of the lists
print("Number of cat images:", len(cat_files))
print("Number of dog images:", len(dog_files))

# Optionally print the first few elements to verify paths
print("First few cat images:", cat_files[:5])
print("First few dog images:", dog_files[:5])

# Check if the list has at least 21 elements before accessing the 21st element
if len(cat_files) > 20:
    # Load the image using matplotlib
    image = plt.imread(cat_files[20])

    # Display the image
    plt.imshow(image)
    plt.axis('off')  # Hide the axis
    plt.show()
else:
    print("The list 'cat_files' does not have enough elements. Current number of images:", len(cat_files))


```

#Image channels


1.   RGB channels




```python


# Display RGB Channels of our image
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img_mpl[:,:,0], cmap='Reds')
axs[1].imshow(img_mpl[:,:,1], cmap='Greens')
axs[2].imshow(img_mpl[:,:,2], cmap='Blues')
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')
plt.show()

```
